```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

!pip install joypy
import joypy as jp

!pip install seaborn_qqplot
from seaborn_qqplot import pplot
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Visualization Project/pokemon.csv')
```

```python
df1 = df.groupby(['type1','generation']).size().unstack()
df1.loc[df1.sum(axis=1).sort_values(ascending = False).index].plot.bar(stacked=True, figsize=(16,8))

plt.xlabel('Primary Type')
plt.ylabel('No. of Pokémon')
plt.title('Primary Type of Pokémon by Generation')
```

```python
sns.set(rc={'figure.figsize':(18,9)})
sns.boxplot(data=df, x='type1', y='base_total')

plt.xlabel('Primary Type')
plt.ylabel('Base Stat Total')
plt.title('Distribution of Total Base Stats based on Primary Type')
```

```python
sns.set(rc={'figure.figsize':(18,9)})
sns.boxplot(data=df, x='type2', y='base_total')

plt.xlabel('Secondary Type')
plt.ylabel('Base Stat Total')
plt.title('Distribution of Total Base Stats based on Secondary Type')
```

```python
sns.stripplot(data=df, x='type1', y='hp', jitter=0.4)

plt.xlabel('Primary Type')
plt.ylabel('HP')
plt.title('Distribution of Health based on Primary Type')
```

```python
sns.displot(df['experience_growth'], kind='kde')
plt.xlabel('Experience Growth')
plt.ylabel('Density')
plt.title('Density of Experience Growth amongst all Pokémon')
```

```python
jp.joyplot(data=df, column='hp', by='generation', ylim='own', legend=True, color='#59B24B')
plt.xlabel('HP Stat')
plt.ylabel('Generation')
plt.title('Density of HP Stat based on Generation')
```
